```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# Load Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)

# Prompt for creative artwork
prompt = """
A futuristic floating city in the clouds, glowing neon lights, flying cars,
giant crystal towers, cinematic lighting, ultra-detailed, fantasy sci-fi art
"""

# Generate image
image = pipe(prompt, guidance_scale=8.5).images[0]

# Show image
plt.imshow(image)
plt.axis("off")

# Save image
image.save("creative_art.png")

print("Image saved as creative_art.png")
```
